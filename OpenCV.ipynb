Face Detection (using test1 image)

```python
# Import the OpenCV library
import cv2
```

```python
# Using the pre-trained classifier from OpenCV
face_cascade=cv2.CascadeClassifier('/content/drive/MyDrive/Colab Notebooks/My Workouts/face_detector.xml') # Load the cascade
```

```python
# Upload the input image
img=cv2.imread('/content/drive/MyDrive/Colab Notebooks/My Workouts/test1.jpg') # Read the input image
```

```python
# Perform face detection
faces=face_cascade.detectMultiScale(img,1.1,4) # Detect faces
```

```python
# Drawing rectangle around the faces
for(x,y,w,h) in faces:
   cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
```

```python
# Exporting the image file
# Export the result
cv2.imwrite('/content/drive/MyDrive/Colab Notebooks/My Workouts/face_detected.png',img)
print('Photo successfully exported!')
```
