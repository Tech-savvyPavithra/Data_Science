```python
d={'col1':[0,1,2,3],'col2':pd.Series([2,3],index=[2,3])}
d
```

```python
df2=pd.DataFrame(data=d,index=[0,1,2,3])
df2
```

```python
df2['newcol']=df2['col1']+df2['col2']
df2
```

```python
df2=df2.append({'col1':1,'col2':2},ignore_index=True)
df2
```

```python
df2.isnull().sum()
```
