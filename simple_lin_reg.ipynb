```python
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('/content/drive/MyDrive/Courses Codes and Datasets/Great Learning/Salary.csv')
df
```

```python
%matplotlib inline
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.scatter(df.YearsExperience,df.Salary,color='green',marker='*')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['YearsExperience']],df.Salary)
```

```python
reg.predict([[14]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
(8731.94104706*14)+28859.52619453144
```

```python
%matplotlib inline
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.scatter(df.YearsExperience,df.Salary,color='green',marker='*')
plt.plot(df.YearsExperience,reg.predict(df[['YearsExperience']]),color='orange') # or reg.predict(YearsExperience)-> easy
```
