```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import networkx as nx

# Create a directed graph
G = nx.DiGraph()

# Add nodes and their positions
nodes = {
    "Data Collection": (0, 2),
    "Real-Time Data Preprocessing": (2, 2),
    "AI Engine": (4, 2),
    "ACO Algorithm": (6, 2),
    "API Integration": (4, 1),
    "Traffic Light Controller": (2, 0),
    "Sensors": (0, 3),
    "GPS": (-2, 2.5),
    "IR Sensor": (-2, 2),
    "Environment Sensor": (-2, 1.5),
    "Cloud Storage": (6, 3),
    "Existing Software": (2, -1),
}

for node, pos in nodes.items():
    G.add_node(node, pos=pos)

# Add edges
edges = [
    ("Sensors", "Data Collection"),
    ("GPS", "Data Collection"),
    ("IR Sensor", "Data Collection"),
    ("Environment Sensor", "Data Collection"),
    ("Data Collection", "Real-Time Data Preprocessing"),
    ("Real-Time Data Preprocessing", "AI Engine"),
    ("AI Engine", "ACO Algorithm"),
    ("ACO Algorithm", "API Integration"),
    ("API Integration", "Traffic Light Controller"),
    ("Traffic Light Controller", "Existing Software"),
    ("AI Engine", "Cloud Storage"),
]

G.add_edges_from(edges)

# Define positions for nodes
pos = nx.get_node_attributes(G, 'pos')

# Draw the graph
plt.figure(figsize=(12, 8))
nx.draw(G, pos, with_labels=True, node_size=5000, node_color='skyblue', font_size=10, font_weight='bold', arrows=True, connectionstyle='arc3,rad=0.1', edge_color='gray')

# Add images at specific nodes (e.g., sensors)
img_path = '/mnt/data/1.jpg'
img = mpimg.imread(img_path)
imagebox = plt.imshow(img)

# Add a title
plt.title("AI-Based Traffic Management System Overview", fontsize=16)

# Show the plot
plt.show()
```
