```python
import pandas as pd

df= pd.read_csv('/content/drive/MyDrive/Computational Statistics/Lab/Exercise 1/iris--.csv')
print(df)
```

```python

plt.figure(figsize = (10, 7))
x = df["SepalLengthCm"]
plt.hist(x, bins = 20, color = "orange")
plt.title("Sepal Length in cm")
plt.xlabel("Sepal_Length_cm")
plt.ylabel("Count")
plt.show()

plt.figure(figsize = (10, 7))
x = df.SepalWidthCm
plt.hist(x, bins = 20, color = "brown")
plt.title("Sepal Width in cm")
plt.xlabel("Sepal_Width_cm")
plt.ylabel("Count")
plt.show()

plt.figure(figsize = (10, 7))
x = df.PetalLengthCm
plt.hist(x, bins = 20, color = "purple")
plt.title("Petal Length in cm")
plt.xlabel("Petal_Length_cm")
plt.ylabel("Count")
plt.show()

plt.figure(figsize = (10, 7))
x = df.PetalWidthCm
plt.hist(x, bins = 20, color = "pink")
plt.title("Petal Width in cm")
plt.xlabel("Petal_Width_cm")
plt.ylabel("Count")
plt.show()
```

```python
new_data = df[["SepalLengthCm"]] #Selected attribute is 'SepalLengthCm'
print(new_data.head())
import matplotlib.pyplot as plt
plt.figure(figsize = (10, 7))
new_data.boxplot()
```

```python
# removing Id column and boxplot for all attributes except Id
new_data = df[["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]]
print(new_data.head())
plt.figure(figsize = (10, 7)) #Boxplot
new_data.boxplot()
```
