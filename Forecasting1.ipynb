```python
# EDA

import pandas as pd # Data Manipulation
from matplotlib import pyplot as plt # Data Visualization
import seaborn as sns
import datetime as dt
import numpy as np # Basic Computation
```

```python
covid= pd.read_csv('/content/drive/MyDrive/Courses Codes and Datasets/Great Learning/covid_19_india.csv',parse_dates=['Date'],dayfirst=True)
```

```python
covid.head()
```

```python
covid=covid[['Date','State/UnionTerritory','Cured','Deaths','Confirmed']]
covid.columns=['date','state','cured','deaths','confirmed']
covid.head(15)
```

```python
covid.tail()
```

```python
today=covid[covid.date=='2021-04-22']
today.shape
today.head()
```

```python
# Sorting data with respect to confirmed cases
max_confirmed_cases=today.sort_values(by='confirmed', ascending=False) # descending
max_confirmed_cases.head(10)
```

```python
top_states_confirmed=max_confirmed_cases[0:10]
top_states_confirmed
```

```python
sns.set(rc={'figure.figsize':(15,8)})
sns.barplot(x="state",y="confirmed",data=top_states_confirmed,hue="state")
plt.show()
```

```python
max_death_cases=today.sort_values(by='deaths', ascending=False) # descending
max_death_cases.head()
```

```python
top_states_death=max_death_cases[0:10]
top_states_death
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x="state",y="deaths",data=top_states_death,hue="state")
plt.show()
```

```python
# Analysis on Cured cases

max_cured_cases=today.sort_values(by='cured', ascending=False) # descending
max_cured_cases.head()
```
