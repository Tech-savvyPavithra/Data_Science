```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradient_descent(x,y):
    m_curr=b_curr=0
    iterations=1000
    n=len(x)
    learning_rate=0.001
    for i in range(iterations):
        y_predicted=m_curr*x+b_curr
        cost=(1/n)*sum([val**2 for val in (y-y_predicted)])
        md=-(2/n)*sum(x*(y-y_predicted))
        bd=-(2/n)*sum(y-y_predicted)
        m_curr=m_curr - learning_rate*md
        b_curr=b_curr-learning_rate*bd
        print("m {},b {},iteration {}".format(m_curr,b_curr,cost,i))

x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])

gradient_descent(x,y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Initialize m_curr and b_curr
m_curr = b_curr = 0

# Your existing code for gradient descent
def gradient_descent(x, y):
    global m_curr, b_curr  # Declare them as global variables
    iterations = 10000
    n = len(x)
    learning_rate = 0.08
    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        cost = (1/n) * sum([val**2 for val in (y - y_predicted)])
        md = -(2/n) * sum(x * (y - y_predicted))
        bd = -(2/n) * sum(y - y_predicted)
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        print(f"m {m_curr}, b {b_curr}, iteration {i}")

# Given data points
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])

# Run gradient descent
gradient_descent(x, y)

# Plot the data points
plt.scatter(x, y, label="Data points")

# Calculate the regression line
x_line = np.linspace(min(x), max(x), 100)
y_line = m_curr * x_line + b_curr
plt.plot(x_line, y_line, color='red', label="Regression line")

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Gradient Descent for Linear Regression")
plt.legend()
plt.grid(True)
plt.show()
```
